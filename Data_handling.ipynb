```python
import fitz  # PyMuPDF

def extract_first_n_pages(pdf_path, n_pages=19):
    doc = fitz.open(pdf_path)
    text = ""
    for page_num in range(min(n_pages, len(doc))):
        text += doc[page_num].get_text()
    return text
```

```python
from pdf2image import convert_from_path
import pytesseract

def extract_text_from_pdf_with_ocr(pdf_path, max_pages=19, lang="ben"):
    images = convert_from_path(pdf_path, first_page=1, last_page=max_pages)
    full_text = ""
    for img in images:
        text = pytesseract.image_to_string(img, lang=lang)
        full_text += text + "\n"
    return full_text
```

```python
import re

def clean_bangla_ocr_text(text):
    # 1. Normalize line breaks to a space (preserves sentence continuity)
    text = text.replace('\n', ' ')

    # 2. Remove multiple consecutive spaces
    text = re.sub(r'\s+', ' ', text)

    # 3. Remove garbage characters (non-Bengali, non-punctuation)
    # Bengali Unicode range: \u0980 to \u09FF
    # Preserve common Bengali punctuation (। and ?)
    text = re.sub(r'[^\u0980-\u09FF\s।?!]', '', text)

    # 4. Strip leading/trailing whitespace
    return text.strip()
```

```python
def chunk_text(text, chunk_size=500, overlap=100):
    chunks = []
    start = 0
    text_len = len(text)

    while start < text_len:
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start += chunk_size - overlap
    return chunks
```

```python
from sentence_transformers import SentenceTransformer

# Bengali-compatible embedding model
model = SentenceTransformer("distiluse-base-multilingual-cased-v2")

def embed_chunks(chunks):
    embeddings = model.encode(chunks, convert_to_tensor=False)
    return list(zip(chunks, embeddings))
```

```python
def process_bangla_text_for_rag(text, chunk_size=500, overlap=100):
    chunks = chunk_text(text, chunk_size, overlap)
    embedded = embed_chunks(chunks)
    return embedded  # List of (chunk, vector)
```
